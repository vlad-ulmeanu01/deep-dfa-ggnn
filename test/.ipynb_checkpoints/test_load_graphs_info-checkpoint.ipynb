```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pickle
import tqdm
import json
```

```python
TEST_FOLDER = "/home/vlad/Desktop/Probleme/deep_dfa_local/local_implementation/dummy_data"
REAL_FOLDER = "/home/vlad/Desktop/Probleme/deep_dfa_local/DDFA/storage/processed/bigvul"

TOP_K_PROP_NAMES = 1000 # cate API/datatype/literal/operator sa tinem (top cele mai frecvente)
```

```python
df_tokens = pd.read_csv(f"{REAL_FOLDER}/abstract_dataflow_hash_api_datatype_literal_operator.csv")
```

```python
df_tokens.shape
```

```python
ht_properties = {"api": {}, "datatype": {}, "literal": {}, "operator": {}}

arr_hashes = df_tokens["hash"]
for hh in tqdm.tqdm(arr_hashes):
    hh = eval(hh)
    for prop in ht_properties:
        for name in hh[prop]:
            if name not in ht_properties[prop]:
                ht_properties[prop][name] = 0
            ht_properties[prop][name] += 1
```

```python
ht_properties
```

```python
[(prop, len(ht_properties[prop])) for prop in ht_properties]
```

```python
sorted_ht_properties = {
    prop: [name for name, cnt in sorted(ht_properties[prop].items(), key = lambda z: -z[1])[:TOP_K_PROP_NAMES]]
    for prop in ht_properties
}
```

```python
sorted_ht_properties
```

```python
with open(f"{TEST_FOLDER}/used_property_names.json", "w") as fout:
    json.dump(sorted_ht_properties, fout)
```

```python
min_max_dgl_id_ht = {}
# min_max_node_id_ht = {}

for dgl_id, node_id, graph_id in zip(list(df_nodes["dgl_id"]), list(df_nodes["node_id"]), list(df_nodes["graph_id"])):
    if graph_id not in min_max_dgl_id_ht:
        min_max_dgl_id_ht[graph_id] = [dgl_id, dgl_id]
    else:
        m, M = min_max_dgl_id_ht[graph_id]
        min_max_dgl_id_ht[graph_id] = [min(dgl_id, m), max(dgl_id, M)]
        
    # if node_id not in min_max_node_id_ht:
    #     min_max_node_id_ht[node_id] = [dgl_id, dgl_id]
    # else:
    #     m, M = min_max_node_id_ht[node_id]
    #     min_max_node_id_ht[node_id] = [min(dgl_id, m), max(dgl_id, M)]
```

```python
for graph_id, node_id in zip(list(df_tokens["graph_id"]), list(df_tokens["node_id"])):
    if graph_id in min_max_dgl_id_ht:
        assert min_max_dgl_id_ht[graph_id][0] <= node_id <= min_max_dgl_id_ht[graph_id][1], f"{graph_id = }, {node_id = }"
```
